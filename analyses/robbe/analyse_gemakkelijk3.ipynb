Bij het wedden in de Play Offs wille we kunnen rekening houden met hoeveel ervaring een club heeft met spelen van de Play Offs

Als de roepnaam van een club gegeven wordt, laten we zien hoeveel seizoenen een club in Play Off 1 en in Play Off 2 heeft gespeeld tot nu toe.

```python
# Importing the necessary packages
import pandas as pd                                 # Data Frame                            # Advanced data visualisation
```

```python
#data inladen
data = pd.read_csv(r"C:\Users\Robbe\Documents\School_jaar_2\sem2\DEP\wedstrijden_playoffs_i_en_ii.csv", sep=",")
```

```python
def count_seasons(df):
    # Voeg de thuis- en uitwedstrijden samen voor een volledig overzicht
    thuis = df[['roepnaam_thuis', 'type', 'seizoen']].rename(columns={'roepnaam_thuis': 'club'})
    uit = df[['roepnaam_uit', 'type', 'seizoen']].rename(columns={'roepnaam_uit': 'club'})
    combined = pd.concat([thuis, uit])
    
    # Groepeer op club en type, en tel unieke seizoenen
    return combined.groupby(['club', 'type'])['seizoen'].nunique().unstack(fill_value=0)

# Bereken het aantal seizoenen voor elke club in elk type Play Off
playoff_season_counts = count_seasons(data)

# Toon de resultaten
playoff_season_counts
```

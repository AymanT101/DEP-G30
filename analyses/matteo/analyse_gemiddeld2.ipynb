Bereken hoeveel je de afgelopen jaren in totaal zou verdiend hebben mocht je telkens consequent 10 EUR hebben ingezet op de laagste van de 3 mogelijke bets (thuis wint / gelijkspel / uit wint)
Als er geen laagste was (bijvoorbeeld bet voor thuis wint = bet voor gelijkspel), dan neem je deze record niet op in de berekeningen.
Voer de berekening eerst uit voor 1 van de 6 aanbieders van weddenschappen (B365 / BS / BW / IW / WH / VC)
Herneem de berekening, maar maak nu gebruik van het maximum bedrag dat je had kunnen krijgen.
Bijvoorbeeld
- B365 --> 1.22 EUR (Bet 365)
- BS --> 1.25 EUR (Blue Square)
- BW --> 1.20 EUR (Bet & Win)
- IW --> 1.28 EUR (Interwetten)
- WH --> 1.25 EUR (William Hill)
- VC --> 1.23 EUR (VC Bet)

In dit geval kies je voor IW want daar krijg je de grootste winst uitbetaald (1.28 EUR) 

```python
# De nodige imports inladen
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Datasets inladen
wedstrijddata = pd.read_csv("D:/Hogent/Visual Studio Code/DEP/DEP1-2023-2024-groep30/transfermarkt/data/correcte_data/wedstrijden.csv", sep=";")
betdata = pd.read_csv("D:/Hogent/Visual Studio Code/DEP/DEP1-2023-2024-groep30/transfermarkt/data/correcte_data/bets_historische_data.csv", sep=",")
```

```python
# De eerste 10 rijen van de dataset bekijken
betdata.head(10)
```

```python
# Functie om de laagste unieke quote te filteren voor B365
def filter_min_unique_b365(row):
    quotes = [row['B365ThuisWint'], row['B365Gelijkspel'], row['B365UitWint']]
    min_quote = min(quotes)  # Zoek de laagste quote
    if quotes.count(min_quote) == 1:  # Als de laagste quote maar 1 keer voorkomt
        return min_quote
    return None

# Bereken het totaal verdiende geld met €10 inzet op de laagste quote van B365
betdata['MinQuoteB365'] = betdata.apply(filter_min_unique_b365, axis=1)
winnings_b365 = betdata['MinQuoteB365'].dropna() * 10 - 10
total_winnings_b365 = winnings_b365.sum()  # Som van gewonnen bedragen minus inzet

# Functie om de maximale van de laagste quotes te filteren
def filter_max_min_quote(row):
    max_quotes = []
    sites = ['B365', 'BW', 'IW', 'WH', 'VC']  # Lijst van sites
    for site in sites:
        site_quotes = [row[f'{site}ThuisWint'], row[f'{site}Gelijkspel'], row[f'{site}UitWint']]
        if all(site_quotes):  # Verifieer dat alle quotes geldig zijn
            max_quote = max(site_quotes)  # Zoek de laagste quote
            max_quotes.append(max_quote)  # Voeg de laagste quote toe aan de lijst
    if max_quotes:  # Controleer of er geldige quotes zijn gevonden
        return max(max_quotes)
    return None

# Voeg de maximale van de laagste quotes toe aan de dataframe
betdata['MaxMinQuote'] = betdata.apply(filter_max_min_quote, axis=1)
max_winnings = betdata['MaxMinQuote'].dropna() * 10 - 10
total_max_min_winnings = max_winnings.sum()  # Bereken het totaal gewonnen bedrag minus inzet

# Print de resultaten
print(f'Totaal gewonnen met B365 laagste inzet: €{total_winnings_b365:.2f}')
print(f'Totaal gewonnen met maximale quote inzet: €{total_max_min_winnings:.2f}')
```

```python
# Grafiek met totale winst van beide gevallen
labels = ['B365', 'Maximale Alle Aanbieders'] # Labels
winnings = [total_winnings_b365, total_max_min_winnings] # Winsten

# Staafdiagram
plt.figure(figsize=(10, 6)) # Grootte van de grafiek
bars = plt.bar(labels, winnings, color=['blue', 'yellow']) # Staafdiagram

# Toevoegen van titel en labels
plt.title('Vergelijking van Totale Gewonnen bedragen') # Titel
plt.xlabel('Weddenschapsstrategie') # X-as label
plt.ylabel('Totaal Gewonnen Bedrag (€)') # Y-as label

# Numerieke labels toevoegen boven elke staaf
for bar in bars:
    yval = bar.get_height() # Hoogte van de staaf
    plt.text(bar.get_x() + bar.get_width()/2, yval, f'€{int(yval):,}', va='bottom', ha='center') # Tekst toevoegen

plt.show()
```
